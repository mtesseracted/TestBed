```python
import re

def getInts(str1):
    if(type(str1) == type("")):
        return [int(s) for s in re.split("\(|,|\)| |\[|\]|_|\.",str1) if s.isdigit()]
        #return [int(s) for s in str1.split() if s.isdigit()]
    else:
        return [""]
```

```python

file_object  = open("questival2017b.txt", "r")
s1 = file_object.readline()
s1 = file_object.readline()
elen = 6
print(type(s1))
print(s1)
print(s1.find(' '))
space1 = s1.find(' ')
space2 = s1[-elen:].find(' ')
s1l = len(s1)
iend = space2+(s1l-elen)
iarr = getInts(s1[iend:-1])
print(space2)
s2 = "\""+s1[:space1]+ "\","+str(iarr[0])+","+str(iarr[1])+",\""+s1[space1+1:iend]+"\""
print(s2)
#print(ord(s1[-1]))
```

```python
fread = open("questival2017c.txt", "r")
fwrite = open("questival2017.csv","w")

i=0
for s1 in fread:
    i+=1
    if s1[0] == 'Q':
        try:
            space1 = s1.find(' ')
            space2 = s1[-elen:].find(' ')
            s1l = len(s1)
            iend = space2+(s1l-elen)
            iarr = getInts(s1[iend:-1])
            print(space2)
            s2 = "\""+s1[:space1]+ "\","+str(iarr[0])+","
            if(iarr[1] == 1):
                s2 += "\"Pic\",\""
            elif(iarr[1] == 2):
                s2 += "\"Vid\",\""
            elif(iarr[1] == 3):
                s2 += "\"Any\",\""
            else:
                s2 += "\"UNK\",\""
            
            s2 += s1[space1+1:iend]+"\"\n"
            fwrite.write(s2)

        except Exception as e:
            print("Exception @"+str(i)+", ERROR: "+str(e))
            print(s1)
    else:
        fwrite.write("\""+s1[:-1]+"\"\n")
        
```
